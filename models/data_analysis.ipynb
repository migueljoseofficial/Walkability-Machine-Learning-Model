```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dirty_df = pd.read_csv("../dataset/heartrate_data.csv")
df = dirty_df.dropna()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Data source'], df['Stress score'], marker='o', linestyle='-', color='b')
plt.title('Comparison of data source and Stress Score')
plt.xlabel('Data source')
plt.ylabel('Stress score')
plt.grid(True)
plt.show()
```
